<h3>FoleyGAN: Описание Генеративной Состязательной Сети генерирования звука для видео<h3/>

![Примет_фоли_для_видео_ряда](img/Примет_фоли_для_видео_ряда.png)

Это второй шаг после AutoFoley в создании генеративно состязательныех сетей для создания аудио дорожки по видео дорожке. Это необходимо в индустрии для участков отснятого хорошего видео, но не качественным или отсутствующей звуковой дорожкой.

При решении этой задачи на данный момент доступно получение качественной звуковой дорожки от специализируемых на этих задачах студиях но за достаточно большую сумму, или получение аудио дорожки из YouTube контента но с не достаточно приемлемым качеством. Данная нейросеть создана как компромис между студийных качественных но очень дорогих аудио записей и некачественного контента.

По проведенному анализу выполненных синтетических дорожек было выявлено, что в 86% случаев опрошенные оценили как качественные. 

Данная сеть генерирует пространственный звук который может быть привязан даже к понарамному видео 360о. Тогда как как даже AutoFoley мог генерировать моно звук.  

![Принципиальная_схема_FolyGAN](img/Принципиальная_схема_FolyGAN.png)

FolyGAN на первом этапе обработки разделен на 2 нейронные сети одна из которых анализирует действия для определенных классов и другая определяет спектрограмму действий классов. После предсказание нейронных сетей поступает с добавлением шума в генеративную сеть которая создает спектрограмму для сравнения дискрименатором со спектрограммой настоящей дорожки. Или поступает в модуль ISTFT преобразования синтетической спектрограммы в аудио дорожку.
