<h3>CenterMask<h3/>

https://arxiv.org/pdf/2004.04446.pdf

Данный алгоритм является алгоритмом одного выстрела для задач пиксельной сегментации. Данный алгорит разделен на две параллельно выполняемых метода. Поиск центров объектов на изображении для получения локальной формы объекта при составлении другим методам глобальной карты значимости объектов.

Свободная якорная обводка и одностадийная сегментационный метод включает простоту, быстроту и точность. Данный алгорит показал скоость 12.3 fps на соревнованиях СОСО и продемонстрировал хорошии показатели скорости и точности. Данный метод легко может быть встроен с другими одностадийными детекторами объектов. 

Локальне формы представлены объектными масками для разных элементов в свободных якорных обводках.
Используя представление центров точек, локальные формы позволяют ветвям предсказания использовать грубые макси и что дает возможность разделяя объекты эффективно в любых пересекаемых другими объектами ситуациях.
Глобальная карта значимости представляет реализацию попиксельного выравнивания. Различия в представлении операций выравнивания непосредственно в сегменатациии. Данный модуль прост, быстр и более точен. Глобальная карта значимости генерирует события схожие семантической сегментации, и возможно эта работа сможет смотивировать одностадийные паноптические сегментации в дальнейшем.


![title](img/CenterMask.png)

Общая линейная последовательность  для CenterMask. 
Состоит из пяти направления расположенные после основной сети. Выходные данные направлений имеют одинаковую высоту (H) и ширину (W) но различные каналы.
С - это число категорий, S - размер и форма вектора. Тепловая карта и Отступ направлений предсказания расположения центра точки. 
Форма и размер напавлений предсказания локальных форм для каждого соответствующего расположения. 
Направление Значимости предсказывает Глобальную карту значимости. Из произведения Локальной формы и часть карты значимости получаем финальную маску для каждого экземпляра.
Для удобной визуализации, вся линейная последовательность сегментации представления только в двух изображенных экземпляров фигур, и Глобальная карта значимости изображена в класс-агностической форме.

![title](img/Таблица_сравнения_алгоритмов.png)

Как видно из таблицы CenterMask предсказывать на достаточно высоком уровне при скорости 25.2 fps (Используя нейронную сеть DLA-34). 

![title](img/Архитектура_направлений_формы_и_размера_для_предсказания_локалной_формы.png)

P представляет карту значений передаваемую неронной сетью. H и W представляет высоту и ширину выдоного сигнала. Размеры каналов формы SxS и в количестве 2 каналов размера. Высота и ширина предсказывают размеры относительно точки. 

![title](img/Сравнение_с_аналогичными_алгоритмами.png)

На изображении представлены результаты различных алгоритмов. Mask R-CNN хорошо обнаруживает объекты, но все еще есть артефакты в генерируемых масках. YOLACT сегментирует с высокой точностью, но пропускает объекты и делает ошибки в некоторых пересекаемых объектах (например у ног на изображении с). PolarMask способен отлечить различные элементы, но маска не обладает точностью полигональной репрезинтации объекта. CenterMask (Ours) способен отлечить пересекаемые объекты хорошо и точно сегментирует маски.

Трудности при использовании могут возникнуть при использовании CenterMask из за использования гиперпараметров α и β в формуле потерь. Так как их необохдимо будет подбирать вручную.

![title](img/Функция_потерь.png)

Достоинством данного алгоритма является возможность параллельного вычисления глобальной карты значимости и локальных форм поотделности. 
Также 25.2 кадров в секунду предоставляет возможность обработку информации в реальном времени и при достаточно высоком уровне точности на данный момент.
